# Importing the libraries required for web scraping 

```python
import numpy as np
import pandas as pd
import requests
import re
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/refrigerators/pr?sid=j9e%2Cabm%2Chzg&page=1'
print(requests.get(url))
page = requests.get(url)
page.text
```

```python
soup = BeautifulSoup(page.text)
soup
```

```python
brand = []
capacity = []
door_type = []
cool_type = []
star_rating = []
digital_inverter = []
rating = []
no_of_rating = []
no_of_review = []
compressor_type = []
built_in_stablizer = []
warranty_on_product = []
warranty_on_compressor = []
price = []
original_price = []
discount = []
for i in range(1,58):
    url = f'https://www.flipkart.com/refrigerators/pr?sid=j9e%2Cabm%2Chzg&page={i}'
    request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
    page = requests.get(url,headers =request_header)
    soup= BeautifulSoup(page.text)
    section = soup.find_all('div',class_='tUxRFH')
    for i in section:
       t = i.find('div',class_="KzDlHZ")
       text = t.text
    # this is for seeing our whole title of refriegerator print(text)
       brands = re.findall('^\w+',text)
       if brands:
         brand.append(brands[0])
       else:
         brand.append(np.nan)
       cap = re.findall('\d+.L',text)
       if cap:
         capacity.append(cap[0])
       else:
         capacity.append(np.nan)
       door = re.findall('\w+.Door',text)
       if door:
         door_type.append(door[0])
       else:
         door_type.append(np.nan)
       cool = re.findall('.Direct.Cool',text)
       if cool:
         cool_type.append(cool[0])
       else:
         cool_type.append(np.nan)
       str_rating = re.findall('\d.Star',text)
       if str_rating:
          star_rating.append(str_rating[0])
       else:
          star_rating.append(np.nan)
       dig_inv = re.findall('(.Digital.Inverter|Smart.Inverter)',text)
       if dig_inv:
          digital_inverter.append('yes')
       else:
          digital_inverter.append('no')
       rate = i.find('div',class_="XQDdHH")
       if rate:
          rating.append(rate.text)
       else:
          rating.append(np.nan)
       r_no = i.find('span',class_="Wphh3N")    # yaha pe no_of_rating and review ka kiya hu 
       if r_no:
         n_rating = re.findall('\d.+.Ratings',r_no.text)
         if n_rating:
            no_of_rating.append(n_rating[0])
         else:
            no_of_rating.append(np.nan)
         n_review = re.findall('&\s(.+)',r_no.text)
         if n_review:
            no_of_review.append(n_review[0])
         else:
            no_of_review.append(np.nan)
       sec_2 = i.find('ul',class_="G4BRas")   # yaha pe maine specification scrap kiya hain 
       if sec_2:
         cmp_type = re.findall('^\w+.\w+.Compressor',sec_2.text)
         if cmp_type:
            compressor_type.append(cmp_type[0])
         else:
            compressor_type.append(np.nan)
         stb_type = re.findall('Built-in.Stabilizer',sec_2.text)
         if stb_type:
            built_in_stablizer.append('yes')
         else:
            built_in_stablizer.append(np.nan)
         warr_prod = re.findall('Stabilizer(\d+\s\w+)',sec_2.text)
         if warr_prod:
            warranty_on_product.append(warr_prod[0])
         else:
            warranty_on_product.append(np.nan)
         warr_com = re.findall('\s\d+.',sec_2.text)
         if warr_com:
           warranty_on_compressor.append(warr_com[0])
         else:
           warranty_on_compressor.append(np.nan)
       prices = i.find('div',class_="Nx9bqj _4b5DiR")
       if prices:
          price.append(prices.text)
       else:
          price.append(np.nan)
       org_price = i.find('div',class_="yRaY8j ZYYwLA")
       if org_price:
          original_price.append(org_price.text)
       else:
          original_price.append(np.nan)
       dis = i.find('div',class_="UkUFwK")
       if dis:
          discount.append(dis.text)
       else:
          discount.append(np.nan)
```

```python
print(len(brand))
print(len(capacity))
print(len(door_type))
print(len(cool_type))
print(len(star_rating))
print(len(digital_inverter))
print(len(rating))
print(len(no_of_rating))
print(len(no_of_review))
print(len(compressor_type))
print(len(built_in_stablizer))
print(len(warranty_on_product))
print(len(warranty_on_compressor))
print(len(price))
print(len(original_price))
print(len(discount))
```

# Making the dataframe of the extracted data and exporting it to a csv file 

```python
data = {
    'Brand':brand,
    'Capacity':capacity,
    'Door_Type':door_type,
    'Cool_type':cool_type,
    'Star_rating':star_rating,
    'Digital_inverter':digital_inverter,
    'Rating':rating,
    'No_of_rating':no_of_rating,
    'No_of_review':no_of_review,
    'Compressor_type':compressor_type,
    'Built_in_stablizer':built_in_stablizer,
    'Product_warranty':warranty_on_product,
    'Compressor_warranty':warranty_on_compressor,
    'Price':price,
    'Original_price':original_price,
    'Discount':discount
}
max_length = max(len(column)for column in data.values())
for key in data:
    data[key]= data[key]+[None]*(max_length - len(data[key]))
data1 = pd.DataFrame(data)
data1.to_csv('refrigerator.csv')
```
